```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import matplotlib as mpl 
mpl.rcParams['figure.dpi'] = 400
import seaborn as sns
```

```python
data = pd.read_csv('../data/processed_data/credit_card_default.csv')
```

```python
print('Tamaño de nuestra tabla: ',data.shape)
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
sns.countplot(x='default payment next month', data=data)
plt.xlabel('Incumplimiento próximo mes')
plt.ylabel('Frecuencia')
plt.title('Distribución del incumplimiento de pago próximo mes')
plt.show()
plt.savefig('../data/img/incumplimiento.png')
```

```python
default_proportion = data['default payment next month'].mean()
print(f"Proporción de incumplimiento de pago: {default_proportion:.2%}")
```

```python
# Relación entre el incumplimiento de pago y el género
sns.countplot(x='SEX', hue='default payment next month', data=data)
plt.xlabel('Género')
plt.ylabel('Frecuencia')
plt.title('Incumplimiento de pago según género')
plt.legend(['No incumplimiento', 'Incumplimiento'])
plt.show()
plt.savefig('../data/img/incumplimiento_por_genero.png')

# Relación entre el incumplimiento de pago y el límite de crédito
sns.boxplot(x='default payment next month', y='LIMIT_BAL', data=data)
plt.xlabel('Incumplimiento de pago')
plt.ylabel('Límite de crédito')
plt.title('Distribución del límite de crédito según el incumplimiento de pago')
plt.show()
plt.savefig('../data/img/dist_in_p_g.png')
```

```python
sns.boxplot(x='default payment next month', y='AGE', data=data)
plt.xlabel('Incumplimiento de pago')
plt.ylabel('Edad')
plt.title('Distribución de la edad según el incumplimiento de pago')
plt.show()
plt.savefig('../data/img/dis_in_of_pay.png')
```

```python
sns.countplot(x='EDUCATION', hue='default payment next month', data=data)
plt.xlabel('Nivel educativo')
plt.ylabel('Frecuencia')
plt.title('Incumplimiento de pago según nivel educativo')
plt.legend(['No incumplimiento', 'Incumplimiento'])
plt.show()

plt.savefig('../data/img/dis_in_of_edu.png')
```

```python
# Visualizar la correlación entre las variables numéricas
plt.figure(figsize=(16, 9))
corr = data.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, fmt=".2f")
plt.show()
plt.savefig('../data/img/heatmap.png')
```
